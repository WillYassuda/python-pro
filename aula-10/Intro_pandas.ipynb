Um DataFrame no pandas representa uma estrutura tabular, ou seja, uma tabela ou planilha.

DataFrames são organizados em colunas (sendo cada coluna uma Série) e podem ser indexados tanto por linhas quanto pelo nome de suas colunas.

É possível criar um DataFrane de duferentes formas:

- A partir de uma lista de listas;
- A partir de um dicionário;
- A partir da leitura de um arquivo .csv, um arquivo .xlsx ou qualquer outra fonte tabular.

### Em vez de usar o pandas para fazer o gráfico, vamos usar o matplotlib puro
Obs.: o pandas usa o matplotlib para fazer os gráficos

```python
# desenhar o mesmo gráfico usando o plt
plt.figure(figsize=(14,4))
plt.scatter('timestamp','high', data=df, c='r')
plt.scatter('timestamp','low', data=df, c='b')
plt.plot('timestamp','close', data=df, c='green')
plt.grid()

plt.title('Variação do dolar')
plt.xlabel('Data')
plt.ylabel('valor')
plt.xticks(rotation=90)

plt.legend(['fechamento', 'preço máximo', 'preço mínimo'])

# plt.show()
plt.savefig('figura1.png')
```

```python
# Um exemplo de como colocar dois gráficos um do lado do outro
fig, ax = plt.subplots(1,2)
ax[0].plot('timestamp','low', data=df, c='b')
ax[1].plot('timestamp','high', data=df, c='r')
plt.show()
```

# Scipy

Biblioteca muito usada por matemáticos, cientistas e engenheiros  
Para mais informações: https://www.scipy.org/  


```python

# Vamos fazer uma simples regressão linear para os valores de fechamento

from scipy.stats import linregress

x = df.index
y = df['close'] # ou y=df.close

stats = linregress(x, y)
m = stats.slope # inclinação ou coeficiente angular da reta 
b = stats.intercept # coeficiente linear ou "onde a curva intercepta o eixo y"
# equação da reta: y = mx + b
```

```python
# Para saber o valor de m e b:
print('O valor de m é {} e o valor de b é {}'.format(m,b)) # outra forma de usar o print
```

```python
# y = mx + b
# Colocando no gráfico:
plt.plot(x, m*x + b, color='red')
plt.show()
```

```python
# dados originais de fechamento em azul e em vermelho a regressão linear. Nesse caso uma reta que ajusta todos os pontos (tendência)
plt.plot(x, m*x + b, color='red') # linha de tendência
plt.scatter(y='close', x=x , data=df) # dados originais
plt.show()
```

```python
# Adicionando 
plt.figure(figsize=(14,4))
plt.scatter('timestamp','high', data=df, c='r')
plt.scatter('timestamp','low', data=df, c='b')
plt.plot('timestamp','close', data=df, c='green')
plt.plot(x, m*x + b, color='yellow')
plt.grid()

plt.title('Variação do dolar')
plt.xlabel('Data')
plt.ylabel('valor')
plt.xticks(rotation=90)

plt.legend(['fechamento', 'regressão linear', 'preço máximo', 'preço mínimo'])

plt.show()
```

```python
import seaborn as sns
```

```python
# correlação entre as colunas numéricas (pandas)
df.corr()
```

```python
# um exemplo de gráfico mapa de calor usando a correlação entre as variáveis
sns.heatmap(df.drop(columns='variacao').corr(), cmap='coolwarm', annot=True) # df.drop -> eliminando a coluna variacao
```
